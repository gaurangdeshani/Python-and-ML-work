 ## ploting histogram used for frequency distribution
 

```python
import random
import numpy as np
sales = [20,25,36,45,12]
wages = np.random.rand(10 )
```

```python
days = [1,2,3,4,5,6,7]
temp = [36,45,41,23,45,22,15]

plt.plot(days, temp, color = 'r' , marker = 'o' , linestyle= '--', linewidth= "1", markersize = "10")
plt.title("surat temperature")
plt.xlabel("days")
plt.ylabel("temperature")
plt.show()
```

```python
days = [1,2,3,4,5,6,7]
temp = [36,45,41,23,45,22,15]

plt.plot(days, temp, color = 'r' , marker = 'o' , linestyle= '--', linewidth= "1", markersize = "10")
plt.title("surat temperature")
plt.xlabel("days")
plt.ylabel("temperature")
plt.grid()
# grid of x can be only seen same for y too
plt.show()
```

```python
plt.hist(wages , rwidth=10, color="m")
plt.show()
```

## subplot

```python
# ploting one entire graph as sub graph
x = [1,2,3,4,5,6]
y=  [4,5,6,7,8,9]

plt.subplot(1,2,1)
plt.plot(x,y)

plt.subplot(1,2,2)
plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [4,5,6,7,8,9]
z = [5,4,7,8,7,8]
t = [8,5,6,4,7,4]

plt.subplot(1,2,1)
plt.plot(x,y)

plt.subplot(1,2,2)
plt.plot(x,y,z,t)
plt.show()
```

## piechart

```python
x = [15,15,22,45]
plt.pie(x)
plt.show()
```

```python
x = [15,15,22,45]
labelx= ["apple", "Bannana", "Grapes", "Orange"]
plt.pie(x, labels= labelx, explode = [0,.2,0,0])
plt.legend()
plt.show()
```

```python
x = [15,15,22,45]
labelx= ["apple", "Bannana", "Grapes", "Orange"]
plt.pie(x, labels= labelx, explode = [0,.2,0,0], shadow=True )
plt.legend()
plt.show()
```
